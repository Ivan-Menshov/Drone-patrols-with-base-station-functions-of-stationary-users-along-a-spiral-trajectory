```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x_user=np.random.uniform(-2,12,100)
y_user=np.random.uniform(-2,12,100)
```

```python
#отрисвка окрудности
#нужно указать центр окружности = (x,y) и радиус окружности = R
def plot_circle(x,y,r):
    angles = np.linspace(0,2*np.pi,500)
    x_cir=x+r*np.cos(angles)
    y_cir=y+r*np.sin(angles)
    plt.plot(x_cir,y_cir)
```

```python
R = 3 # радиус покрытия дрона
```

```python
def traektory(x,y,r):
    angles = np.linspace(0,2*np.pi,100)
    X=[]
    Y=[]
    for i in range(len(angles)):
        x_cir=x+r*np.cos(angles[i])
        y_cir=y+r*np.sin(angles[i])
        X.append(x_cir)
        Y.append(y_cir)
    return X,Y
```

```python
def traektory1(x,y,r):
    angles = np.linspace(2/3*np.pi,8/3*np.pi,100)
    X=[]
    Y=[]
    for i in range(len(angles)):
        x_cir=x+r*np.cos(angles[i])
        y_cir=y+r*np.sin(angles[i])
        X.append(x_cir)
        Y.append(y_cir)
    return X,Y
```

```python
def traektory2(x,y,r):
    angles = np.linspace(4/3*np.pi,10/3*np.pi,100)
    X=[]
    Y=[]
    for i in range(len(angles)):
        x_cir=x+r*np.cos(angles[i])
        y_cir=y+r*np.sin(angles[i])
        X.append(x_cir)
        Y.append(y_cir)
    return X,Y
```

```python
def distance(x1,y1,x2,y2):
    return np.sqrt((x2-x1)**2+(y2-y1)**2)
```

```python
conteyner=[]
conteyner.append(traektory(5,5,4))
conteyner.append(traektory1(5,5,4))
conteyner.append(traektory2(5,5,4))
```

```python
len(conteyner[0][0])
```

```python
screen=[]
covered_users=[]
for j in range(len(conteyner[0][0])):
    a=plt.figure(dpi=100,figsize=(8,8))
    plot_circle(5,5,4)
    for i in range(len(conteyner)):
        plt.plot(conteyner[i][0][j],conteyner[i][1][j],'o',color='red')
        plot_circle(conteyner[i][0][j],conteyner[i][1][j],R)
    i_peresechenie=[]
    i_green=[]
    for i in range(len(x_user)):
        for k in range(len(conteyner)):
            for p in range(k,len(conteyner)):
                if k!=p:
                    if (distance(x_user[i],y_user[i],conteyner[k][0][j],conteyner[k][1][j])<=R) and (distance(x_user[i],y_user[i],conteyner[p][0][j],conteyner[p][1][j])<=R):
                        plt.plot(x_user[i],y_user[i],'.',color='red')
                        i_peresechenie.append(i)
            if (i not in i_peresechenie) and (distance(x_user[i],y_user[i],conteyner[k][0][j],conteyner[k][1][j])<=R):
                plt.plot(x_user[i],y_user[i],'.',color='green')
                i_green.append(i)
        if (i not in i_peresechenie) and (i not in i_green):
            plt.plot(x_user[i],y_user[i],'.',color='red')
            i_peresechenie.append(i)
    covered_users.append(len(i_green))
              
    
    plt.xlim(-2,12)
    plt.ylim(-2,12)
    plt.title('Покрыто '+ str(len(i_green)) + ' из '+ str(len(x_user))+' Time: '+str(j) )
    screen.append(a)
    plt.close()
    print(j)
```

```python
screen_name=[]
for i in range (len(screen)):
    screen[i].savefig(str(i)+'saved_figure.png') 
    screen_name.append(str(i)+'saved_figure.png')
    print(i)
```

```python
import os
import imageio
with imageio.get_writer('mygif.gif', mode='I') as writer:
    for filename in screen_name:
        image = imageio.imread(filename)
        writer.append_data(image)
        print(filename)
```
